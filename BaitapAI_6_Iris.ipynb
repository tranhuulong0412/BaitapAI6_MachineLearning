```python
import pandas as pd 
url ='iris.csv'
data = pd.read_csv(url)
import matplotlib.pyplot as plt 
data.hist()
plt.show
```

```python
correlation = data.corr() #mối tương quan giữa các dữ liệu 
print(correlation)
import seaborn as sns
sns.heatmap(correlation,annot = True)
```

```python
data.plot(kind = 'scatter',x='petal.length',y='petal.width',figsize=(9,9),color = 'black')
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X=pd.DataFrame(data['petal.length']),y=data['petal.width'])
print(model.intercept_)
print(model.coef_)
model.score(X=pd.DataFrame(data['petal.length']),y=data['petal.width'])
```

```python
data.plot(kind = 'scatter',x='petal.length',y='petal.width',figsize=(9,9),color = 'black',xlim=(0,20))
y_pred= model.predict(X=pd.DataFrame(data['petal.length']))
plt.plot(data['petal.length'],y_pred,color='blue')
```

```python
model = linear_model.LinearRegression()
X = pd.DataFrame([data['petal.width'],data['petal.width']**2,data['petal.width']**3,data['petal.width']**4
                  ,data['petal.width']**5]).T
model.fit(X,y=data['petal.length'])
print('Intercept cua phuong trinh la: ',model.intercept_)
print('Coef cua phuong trinh la: ',model.coef_)                         #Bậc 5
print('Do chinh xac cua phuong trinh: ',model.score(X,y=data['petal.length']))
```

```python
import numpy as np
range = np.arange(0,3,0.1)
data.plot(kind = 'scatter', x = 'petal.width', y = 'petal.length', figsize = (9,9), color = 'black')
predictor = pd.DataFrame([range,range**2,range**3,range**4,range**5]).T
y_pred = model.predict(X = predictor)
plt.plot(range, y_pred,color = 'blue')
     
```
